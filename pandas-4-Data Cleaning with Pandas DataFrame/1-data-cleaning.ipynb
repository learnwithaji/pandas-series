```python
import pandas as pd
```

```python
df = pd.read_csv('auto_policy.csv')
df.head()
```

```python
df.rename(columns={'PolicyID': 'PolicyNumber'}, inplace=True)
#df = df.rename(columns={'PolicyID': 'PolicyNumber'})
df.head()
```

```python
#DataFrame.isnull is an alias for DataFrame.isna
df.isnull()
```

```python
#Returns True for rows with any missing values, else False.
#axis=0 will affect rows.
#axis=1 will affect columns.
df.isnull().any(axis=1)
```

```python
# Returns the count of missing (NaN) values in each column.
df.isna().sum()
```

```python
# Get rows with null values in any column
rows_with_null = df[df.isnull().any(axis=1)]

# Set width to a larger value
pd.set_option('display.width', 200)  

print(rows_with_null)
```

```python
# Fill missing PremiumAmount with the mean value
df['PremiumAmount'] = df['PremiumAmount'].fillna(df['PremiumAmount'].mean())
df['PremiumAmount'] = df['PremiumAmount'].round(2)

# Fill missing FuelType with 'Unknown'
df['FuelType'] = df['FuelType'].fillna('Unknown')
print (df)
```

```python
df['CarModel'] = df['CarModel'].str.strip()
#df['CarModel'] = df['CarModel'].str.upper()
#df['CarModel'] = df['CarModel'].str.lower()
rows_without_null  = df[df.notna().all(axis=1)]
print(rows_without_null)
```

```python
df.dropna( inplace=True)
###
```

```python
import numpy as np
df.loc[df['CustomerName'] == 'Manoj K', 'FuelType'] =  'Petrol'
#df['FuelType'] = df['FuelType'].replace('Unknown', 'Petrol')

print(df)
```

```python
df.drop_duplicates(inplace=True)
print(df)
```
